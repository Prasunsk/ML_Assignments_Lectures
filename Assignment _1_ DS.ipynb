You have a matrix $X$ of size (10,5) filled with random values. Print the following :
* Minimum value and index where it occured, for each column
* Maximum value and the index where it occured, for each column
* Minimum value and index where it occured, for each row
* Maximum value and index where it occured, for each row
* Minimum value and index where it occured, for the entire matrix
* Maximum value and index where it occured, for the entire matrix

````
np.random.seed(0)
X = np.random.rand(10,5)
print("X={}".format(X))
for i in range(X.shape[1]):
  lowest=X[:,i].min()
  print("MIN in column %s is %s" %(i,lowest))
  print("Index of MIN is {}".format(np.argmin(X[:,i])))
  largest=X[:,i].max()
  print("MAX in column %s is %s" %(i,largest))
  print("Index of MAX is {}".format(np.argmax(X[:,i])))  

for i in range(X.shape[0]):  
  lowest=X[i,:].min()
  print("MIN in row %s is %s" %(i,lowest))
  print("Index of MIN is {}".format(np.argmin(X[i,:])))
  largest=X[i,:].max()
  print("MAX in row %s is %s" %(i,largest))
  print("Index of MAX is {}".format(np.argmax(X[i,:]))) 

print("Min in the matrix is {}".format(np.min(X)))
print("Index of Min is {}".format(np.argmin(X)))
print("Max in the matrix is {}".format(np.max(X)))
print("Index of Max is {}".format(np.argmax(X)))

````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````
fig = plt.figure()
ax = plt.axes()
x = np.arange(-2*np.pi,2*np.pi,0.01)
plt.subplot(3,1,1)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.legend(["sin(x)","cos(x)"])
plt.xlabel('Radians')
plt.ylabel('Sin(x)')
plt.show()
plt.subplot(3,1,2)
plt.magnitude_spectrum(np.sin(x),Fs=10)     # as sin(x) and cos(x) have the same magnitude spectrum in this question
plt.title('Magnitude Spectrum of sin(x) ')
plt.show()
plt.subplot(3,1,3)
fft3 = np.fft.fft(np.sin(x))
plt.plot(x,fft3.imag)
plt.title('FFT of sin(x)')
plt.show()
````
